```python
#相关矩阵
fig =plt.figure()
fig.add_subplot(1,1,1)
correlation = train.corr()
ax = sns.heatmap(correlation)
ax.set_title('Correlation')
```

```python
k = 10 #与价格SalePrice相关性最高的10个特征
cols = correlation.nlargest(k,'SalePrice').index
print(cols)
correlation02 = correlation.reindex(index=cols,columns=cols)
correlation02
```

```python
#相关矩阵
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.heatmap(correlation02)
ax.set_title('10 x Correlation')
```

```python
#统计缺失值
# train.info()
total = train.isnull().sum()
percent = round(total/len(train)*100,2)
missing_data = pd.concat([total,percent],axis=1)
missing_data.columns = ['total','percent']
missing_data.sort_values(by='percent',ascending=False)
```

```python
#缺失值处理
#We'll consider that when more than 15% of the data is missing, we should delete the corresponding variable
#删除所有缺失值超过15%的特征
train_pre = train.drop(missing_data[missing_data.percent>15].index,axis=1)
print(train_pre.shape)
#删除存在缺失值得所有记录
train_pre.dropna(inplace=True)
train_pre
```
